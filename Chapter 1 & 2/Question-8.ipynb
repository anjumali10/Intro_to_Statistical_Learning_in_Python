```python
import numpy as np
import pandas as pd
```

#### Read the data into python and call the loaded data

```python
college = pd.read_csv('College.csv')
college
```

#### Change the first column name and modify data

```python
college3 = college.rename({'Unnamed: 0': 'College'},axis=1)
college3 = college3.set_index('College')
```

```python
college = college3
college.loc['College']
```

#### Produce a numerical summary

```python
college.describe(include = 'all')
```

#### Produce a scatterplot matrix of the specific columns

```python
columns = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[columns]);
```

```python
from matplotlib import pyplot as plt
```

#### Produce side-by-side boxplots of Outstate versus Private.

```python
fig, ax = plt.subplots(figsize=(5, 5))
college.boxplot('Outstate', by = 'Private', ax = ax);
```

```python
college['Elite'] = pd.cut(college['Top10perc'],[0,0.5,1],labels=['No', 'Yes'])
```

```python
# Number of elite universities
num_elite_uni = college['Elite'].value_counts()
num_elite_uni
```

#### Produce side-by-side boxplots of Outstate versus Elite.

```python
fig, ax = plt.subplots(figsize=(5, 5))
college.boxplot('Outstate', by = 'Elite', ax = ax);
```
