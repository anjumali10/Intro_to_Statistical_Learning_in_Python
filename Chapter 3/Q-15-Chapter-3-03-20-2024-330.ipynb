```python
# Importing libraries and important packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ISLP.models import (ModelSpec as MS,
summarize ,
poly)
import statsmodels.api as sm
```

```python
# Loading dataset
from ISLP import load_data
boston_data = load_data('Boston')
```

```python
boston_data.rename(columns = {'crim': 'crime_rate', 'zn': 'residential_land_zoned', 'indus': 'non-retail_business_acres',
                        'chas': 'charles_river', 'nox': 'nitrogen_oxides', 'rm': 'average_number_rooms', 'age': 'owner-occupied<1940',
                        'dis': 'dis_5_boston_emp_centers', 'rad': 'accessibility_radial_highways', 'tax': 'tax', 'ptratio': 'pupil_teacher_ratio',
                        'lstat': 'lower_status', 'medv': 'median_value'}, inplace = True)
```

## (a) For each predictor, fit a simple linear regression model to predict the response. Describe your results.

### residential_land_zoned to predict crime rate

```python
# Separate the predictor and response variable
design = MS(['residential_land_zoned'])
X = design.fit_transform(boston_data)
y = boston_data['crime_rate']
```

```python
# Fit the model
model = sm.OLS(y, X)
results = model.fit()
```

```python
results.summary()
```

```python
# Create scatter plot to display relationship between variables
ax = boston_data.plot.scatter('residential_land_zoned', 'crime_rate', label = 'Data points')
ax.set_xlabel('residential_land_zoned')
ax.set_ylabel('crime_rate')
ax.legend();
```

## (b) Fit a multiple regression model to predict the response using all of the predictors. Describe your results. For which predictors can we reject the null hypothesis H0 : Î²j = 0?

```python
# Separate the predictor and response variable
design = MS(['residential_land_zoned', 'non-retail_business_acres',
       'charles_river', 'nitrogen_oxides', 'average_number_rooms',
       'owner-occupied<1940', 'dis_5_boston_emp_centers',
       'accessibility_radial_highways', 'tax', 'pupil_teacher_ratio',
       'lower_status', 'median_value'])
X = design.fit_transform(boston_data)
y = boston_data['crime_rate']
# Fit the model
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

* We can reject the null hypothesis for:
  * residential_land_zoned
  * dis_5_boston_emp_centers
  * accessibility_radial_highways
  * median_value
